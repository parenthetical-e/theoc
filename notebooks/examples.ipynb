### Imports

```python
import os
import numpy as np
import pandas as pd
from scipy.signal import welch

from IPython.display import Image
import matplotlib
import matplotlib.pyplot as plt

%matplotlib inline
%config InlineBackend.figure_format = 'retina'

import seaborn as sns
sns.set_style('ticks')

matplotlib.rcParams.update({'font.size': 16})
matplotlib.rc('axes', titlesize=16)

from theoc.oc import oscillatory_coupling
```

# Create dict to hold all the examples

```python
examples = {}
example_dists = {}
example_dists["m"] = list(range(10))
```

# Stimulus and oscillation examples

```python
stim_seed = 303
seed = 49
example = oscillatory_coupling(num_pop=50,
                         num_background=5,
                         t=5,
                         osc_rate=2,
                         f=6,
                         g=2,
                         g_max=8,
                         q=0.25,
                         stim_rate=10,
                         frac_std=0.01,
                         m=10,
                         dt=0.001,
                         stim_seed=stim_seed,
                         seed=seed,
                         save=None)
```

```python
example.keys()
```

```python
fig = plt.figure(figsize=(4, 2))
plt.plot(example["times"], example["rescaled"]["stim_ref"], color="darksalmon")
```

```python
fig = plt.figure(figsize=(4, 2))
plt.plot(example["times"], example["rescaled"]["osc_ref"], color="black")

fig = plt.figure(figsize=(4, 2))
plt.bar(range(10), example["p_y"]["osc_p"], color="black")
```

```python
fig = plt.figure(figsize=(4, 2))
plt.plot(example["times"], example["rescaled"]["stim_p"], color="black")
plt.plot(example["times"], example["rescaled"]["stim_ref"], color="darksalmon")
```

```python
# Add stim_ref, _p and the osc
examples["times"] = example["times"]
examples["osc_ref"] = example["rescaled"]["osc_ref"]
examples["osc_p"] = example["rescaled"]["osc_p"]
examples["stim_ref"] = example["rescaled"]["stim_ref"]
examples["stim_p"] = example["rescaled"]["stim_p"]
```

```python
fig = plt.figure(figsize=(4, 2))
plt.bar(range(10), example["p_y"]["osc_p"], color="black")
plt.bar(range(10), example["p_y"]["stim_ref"], color="darksalmon", alpha=0.3)

example_dists["osc_p"] = example["p_y"]["osc_p"]
example_dists["stim_ref"] = example["p_y"]["stim_ref"]
```

```python
fig = plt.figure(figsize=(4, 2))
plt.bar(range(10), example["p_y"]["stim_p"], color="black")
plt.bar(range(10), example["p_y"]["stim_ref"], color="darksalmon", alpha=0.3)

example_dists["stim_p"] = example["p_y"]["stim_p"]
```
